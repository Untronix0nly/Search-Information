```python
from gensim.corpora.wikicorpus import WikiCorpus
import os
from matplotlib import pyplot as plt
import pandas as pd
from collections import Counter
import numpy as np 
from nltk import word_tokenize
from nltk.stem.snowball import SnowballStemmer 
from sklearn.feature_extraction.text import CountVectorizer
from nltk.corpus import stopwords
import statistics
```

```python
global_params = dict()
global_params["Файловая_статистика"] = dict()
global_params["Файловая_статистика"]["Число_докуменов"] = 40000
global_params["Файловая_статистика"]["Максимальное_число_слов"] = -1
global_params["Файловая_статистика"]["Минимальное_число_слов"] = 99999
global_params["Файловая_статистика"]["Общее_число_слов"] = 0
```

```python
def make_wiki_corpus(in_file:str,out_file_pattern:str)->(list,list):
    """
    Парсит указанный архив(дамп), собирает общую статистику и записывает все в файл
    --------------
    Params
    -------
        in_file:str
            Путь к файлу(архиву),содержимое которого необходимо распарсить
        out_file: str
            Путь в файлу для записи распарсенного содержимого
    Returns
    -------
        all_words:list
            Список слов, содержащийся во всех документах
        statistic_dict:dict
            Словарь со статистикой по документам
            
    Exceptions
    ----------
        FileNotFoundError
    """
    if not os.path.exists(in_file):
        raise FileNotFoundError("Файла в указанной директории нет")
        
    wiki = WikiCorpus(in_file,dictionary = False)
    texts = list()
    documents_len = list()
    index = 0
    for article in wiki.get_texts():
        out_file = "{0}{1}.txt".format(out_file_pattern,index)
        with open(out_file,"w",encoding = "utf-8") as stream:
            stream.write(" ".join(article))
        texts.append(" ".join(article)) 
        index+=1
        if index == global_params["Файловая_статистика"]["Число_докуменов"]:
            break
        documents_len.append(len(article))
    return texts,documents_len
    
def freq_dist(sentence,n_grams):
    ngrams_dict = dict()
    for index in range(n_grams,n_grams + 1):
        ngram_vectorizer = CountVectorizer(analyzer = 'word',tokenizer = word_tokenize,ngram_range = (index,index),min_df = 1)
        X = ngram_vectorizer.fit_transform(sentence.split('\n'))
        vocab = list(ngram_vectorizer.get_feature_names())
        counts = X.sum(axis=0).A1
        freq_distribution = Counter(dict(zip(vocab,counts)))
        ngrams_dict["{0} n-grams".format(index)] = freq_distribution
    return ngrams_dict

def plot_statistics_ngrams(texts,n_grams):
    stemmer = SnowballStemmer('russian')
    '''
    for text in texts:
        text = " ".join([stemmer.stem(word) for word in text.lower().split(" ")]) + "\n"
    '''
    for text in texts:
        text = text + '\n'
    text_matrix = " ".join(texts)
    freq_dict = freq_dist(text_matrix,n_grams) 
    titles = freq_dict["{0} n-grams".format(n_grams)].most_common(100)
    print(pd.DataFrame.from_dict(titles))
    stop_words = stopwords.words('russian')
    stop_words.extend(['что', 'это', 'так', 'вот', 'быть', 'как', 'в', '—', 'к', 'на'])
    sets = [title for title in titles if title[0] not in stop_words]
    print(pd.DataFrame.from_dict(sets))
    X,y = [set0[0] for set0 in sets[:10]],[set1[1] for set1 in sets[:10]]
    plt.scatter(X,y)
    plt.title("{0} gram distribution".format(n_grams))

def plot_file_statistics(documents_len):
    global_params["Файловая_статистика"]["Максимальное_число_слов"] = max(documents_len)
    global_params["Файловая_статистика"]["Минимальное_число_слов"] = min(documents_len)
    global_params["Файловая_статистика"]["Общее_число_слов"] = sum(documents_len)
    global_params["Файловая_статистика"]["Медиана"] = statistics.median(documents_len)
    global_params["Файловая_статистика"]["Мода"] = statistics.mode(documents_len)
    global_params["Файловая_статистика"]["Средний объем"] = int(global_params["Файловая_статистика"]["Общее_число_слов"] / global_params["Файловая_статистика"]["Число_докуменов"])
    print(pd.DataFrame.from_dict(global_params))
    plt.hist(documents_len,bins= 20)
    plt.title("Распределение объема текста")
    plt.xlabel("Размер документа")
    plt.ylabel("Число документов")

def plot_total_statistics(sizes):
    global_params["Файловая_статистика"]["Минимальный_размер_файла"] = min(sizes)
    global_params["Файловая_статистика"]["Максимальный_размер_файла"] = max(sizes)
    global_params["Файловая_статистика"]["Общий_размер_файла"] = sum(sizes)
    print(pd.DataFrame.from_dict(global_params))
    plt.hist(sizes,bins= 100)
    plt.title("Распределение объема текста")
    plt.xlabel("Размер документа")
    plt.ylabel("Число документов")
    
            
            
```

```python
words,counts = make_wiki_corpus("data/ruwiki-20181120-pages-articles-multistream.xml.bz2","data/text/wiki")
```
